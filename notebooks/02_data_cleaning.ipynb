```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/telco_churn.csv")
```

```python
df.isnull().sum()
```

```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df["TotalCharges"].isnull().sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)
```

```python
df.drop(
    "customerID",
    axis=1,
    inplace=True
)
```

```python
df.to_csv(
    "../data/processed/cleaned_telco.csv",
    index=False
)
```
