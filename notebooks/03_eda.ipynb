```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/cleaned_telco.csv")
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    x="Churn",
    data=df
)

plt.show()
```

```python
sns.countplot(
    x="Contract",
    hue="Churn",
    data=df
)

plt.xticks(rotation=20)

plt.show()
```

```python
sns.boxplot(
    x="Churn",
    y="MonthlyCharges",
    data=df
)
```
