```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "../data/processed/cleaned_telco.csv"
)
```

```python
churn_rate = (
    (df["Churn"]=="Yes").mean()
    *100
)

print(
    f"Churn Rate: {churn_rate:.2f}%"
)
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    x="Contract",
    hue="Churn",
    data=df
)

plt.title(
    "Contract Type vs Churn"
)

plt.show()
```
