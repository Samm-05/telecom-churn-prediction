```python
# ==========================================================
# TELECOM CHURN PREDICTION
# NOTEBOOK 08 - BUSINESS INSIGHTS
# ==========================================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# ==========================================================
# LOAD DATASET
# ==========================================================

df = pd.read_csv("../data/processed/cleaned_telco.csv")

print("="*50)
print("DATASET INFORMATION")
print("="*50)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

display(df.head())

# ==========================================================
# CHURN RATE
# ==========================================================

churn_rate = (
    (df["Churn"] == "Yes").mean()
    * 100
)

print("\n")
print("="*50)
print("BUSINESS KPI")
print("="*50)

print(f"Overall Churn Rate: {churn_rate:.2f}%")

# ==========================================================
# CONTRACT TYPE VS CHURN
# ==========================================================

plt.figure(figsize=(10,6))

sns.countplot(
    x="Contract",
    hue="Churn",
    data=df
)

plt.title("Contract Type vs Customer Churn")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# ==========================================================
# MONTHLY CHARGES VS CHURN
# ==========================================================

plt.figure(figsize=(10,6))

sns.boxplot(
    x="Churn",
    y="MonthlyCharges",
    data=df
)

plt.title("Monthly Charges vs Customer Churn")
plt.tight_layout()
plt.show()

# ==========================================================
# TENURE VS CHURN
# ==========================================================

plt.figure(figsize=(10,6))

sns.boxplot(
    x="Churn",
    y="tenure",
    data=df
)

plt.title("Customer Tenure vs Churn")
plt.tight_layout()
plt.show()

# ==========================================================
# INTERNET SERVICE VS CHURN
# ==========================================================

plt.figure(figsize=(10,6))

sns.countplot(
    x="InternetService",
    hue="Churn",
    data=df
)

plt.title("Internet Service vs Customer Churn")
plt.tight_layout()
plt.show()

# ==========================================================
# PAYMENT METHOD VS CHURN
# ==========================================================

plt.figure(figsize=(12,6))

sns.countplot(
    x="PaymentMethod",
    hue="Churn",
    data=df
)

plt.xticks(rotation=25)
plt.title("Payment Method vs Customer Churn")
plt.tight_layout()
plt.show()

# ==========================================================
# BUSINESS INSIGHTS
# ==========================================================

insights = [
    "Customers with Month-to-Month contracts churn more frequently.",
    "Customers with higher Monthly Charges have higher churn probability.",
    "Long-tenure customers are less likely to churn.",
    "Electronic Check users show the highest churn rate.",
    "Fiber Optic customers churn more frequently than DSL customers."
]

print("\n")
print("="*50)
print("BUSINESS INSIGHTS")
print("="*50)

for i, insight in enumerate(insights, start=1):
    print(f"{i}. {insight}")

# ==========================================================
# SAVE BUSINESS REPORT
# ==========================================================

os.makedirs("../reports", exist_ok=True)

business_summary = pd.DataFrame({
    "Insight": insights
})

business_summary.to_csv(
    "../reports/business_summary.csv",
    index=False
)

print("\n")
print("="*50)
print("REPORT STATUS")
print("="*50)

print("business_summary.csv saved successfully!")

# ==========================================================
# PROJECT STATUS
# ==========================================================

print("\n")
print("="*50)
print("NOTEBOOK 08 COMPLETED SUCCESSFULLY")
print("="*50)
```
